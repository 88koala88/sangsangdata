cf) 웹 스크래핑과 웹 크롤링의 차이점

### 웹 기본 지식 이해하기

웹이란 월드 와이드 웹의 줄임말이며 인터넷에 연결된 컴퓨터를 통해 사람들이 정보를 공유할 수 있는 전 세계적인 정보 공간을 의미한다. 동시에 그림이나 문자와 같은 데이터를 교환하는 전송 방식을 뜻하기도 한다. 이러한 정보를 열람할 수 있도록 웹 브라우저도 만들어졌는데 엣지, 크롬, 사파리 등이 이에 해당한다. 그리고 다양한 문자와 그림들을 웹 브라우저에서 볼 수 있도록 표현하기 위해 사용하는 기술이 HTML이다.
HTML은 웹 페이지 표시를 위해 개발된 표준 마크업 언어이다. 그 중 HTML의 가장 기본적인 태그를 살펴보자.

- 태그 : 태그는 HTML 문서를 이루는 문법적인 표시로, 문자를 꺾쇠 기호(<>)로 감싼 형태로 작성한다. 대부분 시작 태그와 종료 태그로 구성되고 그 사이에 내용이 들어간다. 시작 태그는 꺾쇠 기호만으로 작성하고, 종료 태그는 문자 앞에 슬래시 기호(/)가 들어간다. 쌍으로 구성된 태그의 경우 시작 태그를 사용했으면 반드시 종료 태그로 닫아야 한다.   ex) <html> ... </html>

- <body> : body 태그는 HTML 문서의 내용 중 실제로 웹 브라우저에 표기할 내용을 담는다. <html> 태그가 열린 다음에 등장하기 때문에 </html> 태그로 닫히기 전에 </body> 태그로 먼저 닫아야 한다.

- < h1 >, < p > : h1 태그는 가장 큰 제목을 표현하기 위한 태그이고, p는 일반 단락을 표기하는 태그이다. 

이처럼 웹 브라우저는 HTML 태그로 구성된 텍스트 파일을 서버에서 다운로드해 사용자에게 보여주는 역할을 한다. 따라서 웹 브라우저에서 HTML 파일을 확인해보고 싶다면 '개발자 도구'를 이용하면 된다. 특히 개발자 도구의 [Elemnets] 탭에서는 웹 페이지를 구성하고 있는 HTML 태그 정보를 쉽게 확인할 수 있고 이를 토대로 웹 스크래핑을 수행할 때 수집할 정보를 가리키는 지표가 되므로 중요하다.

cf) CSS : 웹 페이지에 보이는 스타일을 지정하는 것이 바로 CSS 이다.

웹에서 데이터를 주고받는 데는 HTTP라는 프로토콜이 사용된다. 프로토콜은 인터넷을 통해 데이터를 주고 받기 위해 미리 정의한 약속, 즉 일종의 규약을 의미한다. 따라서 데이터 수집 대상의 요청은 http:// 혹은 https://로 시작한다. 이 주소는 웹 브라우저 주소창에 이볅하는 URL 주소와 같다.
웹 브라우저는 사용자가 HTTP 프로토콜을 사용한 요청을 서버로 전달하고 응답을 받아 화면에 출력하는 역할을 하는 것이다. 
특히 HTTP 요청을 보낼 때 함께 전송하는 데이터를 페이로드라고 한다. 이 데이터는 웹 스크래핑을 수행할 때 어떤 데이터를 함께 보내야 할지 이해하는데 도움이 된다. 또한 서버가 웹 브라우저로 보낸 응답 내용을 Response라고 한다. 
그리고 웹 스크래핑을 하려면 웹 요청을 웹 브라우저가 아니라 주로 파이썬 소스 코드로 보내야 한다. 또한 요청을 보내는 주체가 무엇인지 설명하는 user-agent 값이 중요하므로 이 또한 잘 알아두자.

### 데이터 수집 및 가공

이번 프로젝트에서는 각종 금융 관련 지표를 쉽게 확인할 수 있는 네이버 금융을 이용해보자. 인기 종목 중 하나인 '삼성전자'를 선택해서 '일별 시세' 데이터를 확인해보자. 일별 시세만 업데이트 하는 URL을 찾기 위해 개발자 도구를 이용하자. 
sise_day 항목을 검색해서 해당 일별 시세 페이지 URL 주소를 얻어놓자. (https://finance.naver.com/item/sise_day.naver?code=005930&page=1)
참고로 웹 브라우저에서 데이터를 확인할 수 있다고 해서 무조건 웹 스크래핑을 할 수 있는 것은 아니다. 웹 사이트에서 웹 브라우저가 아닌 방법으로 접근하는 요청을 막는 경우가 있기 때문이다. 따라서 데이터 수집용 URL이 파이썬 소스 코드에서 사용할 수 있는 주소인지 먼저 확인해야 한다.

```python
# 웹 스크래핑 가능 유무 확인을 위한 모듈 탑재
import requests

# 웹 스크래핑 대상 주소 지정 (문자열 처리 해줘야함)
url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page=1'

# HTTP 요청을 대상 주소로 보내고, HTTP 응답 받아서 저장하기
page = requests.get(url)

# 응답 내용 확인하기
print(page.text)
```

위의 내용을 확인해보면 웹 브라우저에서는 아무 문제가 없던 URL이지만, 파이썬 소스 코드에서는 페이지를 찾을 수 없다는 에러 메시지가 뜬다. 이는 네이버 주식 웹 사이트에서 웹 브라우저가 아닌 방식으로 HTTP 요청을 보내면 HTTP 응답을 전송하지 못하게 막아 놓았다는 의미이다. 
그렇다면 이런 경우에는 웹 스크래핑을 할 수 없는 것일까? 그렇지 않다. 이는 소스 코드에 웹 브라우저에 요청할 때 설정한 값을 입력하지 않았기 때문에 발생한 문제이다. 따라서 '웹 브라우저에서 보낸 요청이다'라는 의미의 user-agent 값을 요청 헤더에 설정하면 이 문제를 해결할 수 있다. requests.get() 메서드에 headers 인수를 추가해 user-agent를 설정하고 소스 코드를 실행해보자.

```python
# 헤더 정보 설정하기
my_headers = {'user-agent' : 'Mozilla/5.0'}

# 요청 헤더 설정하여 HTTP 응답 받아오기
page = requests.get(url, headers=my_headers)

# 응답 내용 확인하기
print(page.text)
```

우리가 수집할 페이지의 데이터는 표의 형태, 즉 table로 되어 있으므로 <table> 태그에 있는 값을 가져오는 방법을 사용하자. 이는 판다스의 read_html() 메서드를 사용하면 된다. 인수에는 앞서 HTTP 응답 내용을 확인할 때 사용했던 HTML 페이지의 텍스트(page.text)를 넣으면 된다.

```python
# 판다스 라이브러리 탑재
import pandas as pd

# HTML 페이지에서 테이블 추출
pages = pd.read_html(page.text)

# 추출한 데이터 타입 확인하기
type(pages)
```

위에서 확인할 수 있듯이 read_html() 메서드가 반환한 객체의 데이터 타입은 리스트이다. read_html() 메서드는 HTML 테이블을 읽어 데이터프레임 객체로 구성된 리스트를 반환한다.
그럼 이 리스트에는 몇 개의 데이터프레임 객체가 들어 있는지 파이썬 len() 함수로 확인해보자.

```python
# 페이지 개수 확인하기
len(pages)
```

```python
# 첫번째 테이블 내용 확인하기
pages[0]
```

```python
# 두번째 테이블 내용 확인하기
pages[1]
```

첫번째 항목은 우리가 찾던 데이터이고, 두번째 항목은 테이블 밑에 있는 페이지 바이다. 이처럼 눈에 보이는 테이블 이외에도 HTML에 <table> 태그가 쓰인 곳이 여러 곳일 수 있다. 어떤 항목의 데이터프레임이 원하는 데이터인지 반드시 확인해야 한다. 

데이터를 여러 페이지로 나누어 출력하도록 설정하는 것을 페이징 처리라고 한다. 너무 많은 데이터를 한번에 보여주려 하면 페이지 로딩 시간이 길어질 수 있으며, 자원이 부족해 오류가 발생할 수도 있기 때문에 한 페이지에 제한된 개수의 데이터만 보여주는것이 일반적이다. 
지금까지 사용한 데이터는 첫번째 페이지에 있는 데이터만이었다. 하지만 우리는 그 이상의 여러 페이지에 담긴 데이터가 필요하다. 
방법은 URL을 확인하면 알 수 있다. 우리가 사용한 위의 해당 URL을 살펴보면
https://finance.naver.com/item/sise_day.naver?code=005930&page=1
code=005930&page=1 부분을 확인하면 된다.
URL에는 정적인 주소 외에 동적으로 변경할 수 있는 정보를 메서드 인수와 같이 추가하는 방법이 담겨 있다. 이 방법을 쿼리 문자열 혹은 URL 매개변수(파라미터)라고 부른다. 쿼리 문자열은 URL 끝에서 물음표 기호(?)로 시작하며, 매개변수 이름과 값을 이름 = 값 형식으로 표기한다. 매개변수가 여러개인 경우 엠퍼샌드(&) 기호를 사용해 구분한다. 즉 위의 URL은 2개의 매개변수를 보내고 있다.
code=005930는 현재 선택한 종목을 의미하며 page=1은 현재 출력하고 있는 페이지를 의미한다. 따라서 여러 페이지 정보를 가져오려면 매개변수 페이지 숫자를 증가하면서 데이터를 축적하면 된다.


그럼 페이지 번호를 증가시키는 반복문을 작성하면 될 것이다. 먼저 페이지 번호를 제외한 나머지 부분을 저장한 새로운 변수 new_url을 만들자. 그리고 반복문을 순회하면서 추출한 테이블 정보를 축적할 데이터프레임인 all_tables를 만들자.

```python
# 페이지 번호를 제외한 주소 저장
new_url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page='

# 데이터를 축적할 데이터프레임 생성
all_tables = pd.DataFrame()
```

이제 페이지 번호 증가를 위한 for문을 작성해보자.

```python
# 페이지 번호 증가용 반복문
for page_number in range (1, 2):
    
    # 페이지 번호 추가한 주소 완성
    full_url = new_url + str(page_number)
    
    # 주소 확인하기     --> 이런 텍스트를 '로그'라고 부른다.
    print(f'{page_number}번째 페이지 읽어오기({full_url})')
    
    # HTTP 요청 전송 후 응답 받아오기
    page = requests.get(full_url, headers=my_headers)
    
    # 테이블 추출
    table = pd.read_html(page.text)[0]
    
    # 수행할 내용 확인
    print(f'전체 {len(all_tables.index)} 줄에 {len(table.index)} 줄 추가')
    
    # 데이터 축적용 데이터프레임에 테이블 추가      
    # concat()메서드를 사용해 이번 페이지 정보를 붙여넣는다. 인수에 리스트가 위치. 리스트 항목에 들어가는 변수 순서에 유의!
    all_tables = pd.concat([all_tables, table])
    
```

위의 for문을 1회 수행한 결과를 확인해보자. 

```python
# 전체 테이블 결과 확인
all_tables
```

위의 결과를 보면 결과에 NaN으로 채워진 줄이 보인다. NaN은 Not a Number라는 의미로, 값이 존재하지 않는 경우에 나타난다. 이 값은 HTML <table>태그 안에 존재하는 행이지만 데이터는 포함하고 있지 않는다. 눈에 보이는 테이블 테두리나 날짜 구분을 위해 추가된 행이므로 숫자가 보이지 않는 것이다. 필요없는 결손치이므로 제거해야한다. dropna()메서드를 사용하자.

```python
# 결손치 제거
all_tables.dropna(inplace=True)

all_tables
```

이제 10페이지를 모아 100일 이상의 데이터를 수집해보자. for문의 range()인수가 1부터 10까지 순회하도록 수정하고 실행하자.

```python
# 데이터를 축적할 데이터프레임 생성
all_tables = pd.DataFrame()

# 페이지 번호 증가용 반복문
for page_number in range (1, 11):
    
    # 페이지 번호 추가한 주소 완성
    full_url = new_url + str(page_number)
    
    # 주소 확인하기   
    print(f'{page_number}번째 페이지 읽어오기({full_url})')
    
    # HTTP 요청 전송 후 응답 받아오기
    page = requests.get(full_url, headers=my_headers)
    
    # 테이블 추출
    table = pd.read_html(page.text)[0]
    
    # 수행할 내용 확인
    print(f'전체 {len(all_tables.index)} 줄에 {len(table.index)} 줄 추가')
    
    # 데이터 축적용 데이터프레임에 테이블 추가      
    all_tables = pd.concat([all_tables, table])
    
```

이제 결손치를 제거하고 수집한 전체 데이터를 확인해보자.

```python
# 결손치 제거
all_tables.dropna(inplace=True)

all_tables
```

### 데이터 시각화

주식 시세 그래프를 확인할 때 항상 등장하는 그래프가 바로 '상자 수염 그래프'이다. 영어로 간단히 '박스 플랏'이라고도 한다. 상자 수염 그래프는 지금까지 다뤘던 그래프와는 다르게 다음과 같이 데이터를 요약한 다섯 가지로 지표를 표기한다.

- 최대값 : 가장 큰 값
- 제 3사분위 : 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분위 : 중앙값
- 제 1사분위: 전체 데이터 중 상위 25%에 해당하는 값
- 최솟값 : 가장 작은 값

이 그래프는 데이터프레임의 boxplot() 메서드를 사용하면 쉽게 그릴 수 있다. 여기에서는 일일 시세 데이터를 월별로 취합해 상자 수염 그래프에 다섯 가지 요약 수치를 표현하는 방식으로 시각화를 진행해보자.

그런데 한글을 그래프로 표현하면 폰트에 따라 그래프 라벨이 깨져서 보이는 경우가 있다. 또한 데이터 처리시 문자 세트 불일치로 인해 문제가 생기기도 한다. 데이터를 안전하게 처리하고 시각화하기 위해 다음과 같이 데이터프레임의 rename() 메서드를 사용해 필요한 열 이름만 영문으로 변경해보자.

- 날짜 : date
- 종가 : end_price
- 시가 : start_price
- 고가 : highest_price
- 저가 : lowest_price

```python
# 열 이름 영문으로 변경
all_tables.rename(columns={'날짜':'date',
                           '종가':'end_price',
                           '시가':'start_price',
                           '고가':'highest_price',
                           '저가':'lowest_price'}
                 , inplace=True)

all_tables
```

위에서 영문으로 표기하지 않은 전일비와 거래량은 필요가 없는 열이다. 데이터프레임의 drop() 메서드로 제거하자.

```python
# 필요없는 열 제거
all_tables.drop(['전일비','거래량'], axis=1, inplace=True)

all_tables
```

상자 수염 그래프에 표기할 데이터는 최고가와 최저가의 중앙값으로 한다. 중앙값을 계산해 추가해보자.

```python
# 중앙값 열 추가
all_tables['midian_price']=all_tables['highest_price'] - ((all_tables['highest_price'] - all_tables['lowest_price'])/2)

all_tables
```

일일 시세를 월별로 집계하려면 월 정보가 필요하다. date 열에 있는 일자 정보를 문자열로 변경해 파이썬 슬라이싱으로 월까지 내용을 잘라서 추가해보자.

```python
# 월 정보 열 추가
all_tables['month'] = all_tables['date'].str[0:7]

all_tables
```

이제 그래프 x축을 숫자로 설정하기 위해 데이터프레임의 set_index() 메서드를 사용해 색인을 date로 변경하자. 그리고 시간상 역순으로 되어 있는 데이터를 색인 오름차순으로 정렬해 왼쪽에서 오른쪽으로 날짜가 증가하는 형태로 그래프를 그리자. 정렬은 데이터프레임의 sort_index()를 사용한다.

```python
# 날짜를 색인으로 지정하기
all_tables.set_index('date', inplace=True)

# 날짜 오름차순으로 정렬하기
all_tables.sort_index(inplace=True)

all_tables
```

시각화를 위한 데이터 2차 가공이 모두 끝났다. 따라서 이제 시각화를 진행해보자.
데이터가 준비되면 먼저 가장 기본 그래프인 선 그래프를 그려 데이터를 살펴보는 것이 좋다.

```python
# 전체 숫자 데이터 선 그래프로 그리기
all_tables.plot.line()
```

위의 그래프를 보면 그래프 간격이 잘 보이지 않는 구간이 존재한다. 데이터를 조금 더 이해하기 위해 앞 15개 정도만 다시 확인해보자.
슬라이싱으로 앞 15개 데이터만 선택하고 x축 라벨은 45도 돌려서 표현해보자.

```python
# 앞 15일치 데이터 선 그래프로 그리기
all_tables[:15].plot.line(rot=45)
```

이제 상자 수염 그래프를 그려보자. 전체 데이터를 월 단위로 집계하여 중앙값의 월 단위 최대값, 최소값, 중앙값 등을 확인하자. 
그래프를 그릴 열을 선택하기 위해 인수 column에 midian_price 열을 설정해주고, 월 단위로 데이터를 집계하기 위해 by 인수에 month 열의 값을 설정해주자.

```python
# 일일 시세 중앙값의 상자 수염 그래프 그리기
all_tables.boxplot(column=['midian_price'], by=['month'])
```
